```python
import cv2
from skimage import feature
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Baca citra
image_nyala = cv2.imread('res/gelap/nyala/175-0.jpg')
image_mati = cv2.imread('res/gelap/mati/440-0.jpg')

# Konversi ke citra grayscale
gray_image_nyala = cv2.cvtColor(image_nyala, cv2.COLOR_BGR2GRAY)
gray_image_mati = cv2.cvtColor(image_mati, cv2.COLOR_BGR2GRAY)


# Buat subplot dengan 1 baris dan 2 kolom
fig, axes = plt.subplots(1, 2)

# Tampilkan gambar pertama pada kolom pertama
axes[0].imshow(gray_image_nyala, cmap='gray')
axes[0].set_axis_off()
axes[0].set_title('#1')

# Tampilkan gambar kedua pada kolom kedua
axes[1].imshow(gray_image_mati, cmap='gray')
axes[1].set_axis_off()
axes[1].set_title('#2')

# Tampilkan canvas dengan kedua gambar
plt.show()
```

```python
# Hitung LBP
radius = 1
n_points = 8 * radius
lbp_image_nyala = feature.local_binary_pattern(gray_image_nyala, n_points, radius, method="uniform")
lbp_image_mati = feature.local_binary_pattern(gray_image_mati, n_points, radius, method="uniform")

# print(lbp_image_nyala)
# print(lbp_image_mati)
```

```python
lbp_hist_nyala, _ = np.histogram(lbp_image_nyala.ravel(), bins=np.arange(0, n_points + 3), range=(0, n_points + 2))
lbp_hist_nyala = lbp_hist_nyala / (lbp_hist_nyala.sum() + 1e-6)


lbp_hist_mati, _ = np.histogram(lbp_image_mati.ravel(), bins=np.arange(0, n_points + 3), range=(0, n_points + 2))
lbp_hist_mati = lbp_hist_mati / (lbp_hist_mati.sum() + 1e-6)
```
