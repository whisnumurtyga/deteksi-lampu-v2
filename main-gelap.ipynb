```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.image import imread
```

```python
dir_on = {
    'beat':'dataset-baru/beat-gelap-nyala', 'cbr':'dataset-baru/cbr-gelap-nyala',
}

dir_off = {
    'beat':'dataset-baru/beat-gelap-mati', 'cbr':'dataset-baru/cbr-gelap-mati',
}
```

```python
beat_on = f"{dir_on['beat']}/2.jpg"
beat_off = f"{dir_off['beat']}/2.jpg"
image_paths = [beat_on, beat_off]
```

```python
# Buat jendela gambar
plt.figure(figsize=(12, 4))

# Loop melalui path file gambar dan menampilkannya
for i, image_path in enumerate(image_paths):
    # print(image_path)
    plt.subplot(1, len(image_paths), i + 1)  # 1 baris, n kolom
    image = imread(image_path)
    plt.imshow(image)
    plt.axis('off')  # Matikan tanda sumbu x dan y

# Tampilkan semua gambar dalam satu jendela tampilan
plt.show()
```

```python
img_beat_on = imread(beat_on)
img_beat_off = imread(beat_off)

plt.imshow(img_beat_on)
plt.axis('off') 
plt.show()

plt.imshow(img_beat_off)
plt.axis('off') 
plt.show()
```
